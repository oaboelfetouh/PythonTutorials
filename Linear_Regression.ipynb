## Linear Regression
Linear regression is a statistical modeling technique used to analyze the relationship between a dependent variable (y) and one or more independent variables (x1, x2, x3, ...). It aims to find a linear equation (y = w1.x1 + w2.x2 + w3x3 + .... + c) that best represents the relationship between the variables.

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
import random

random.seed(42)  # Set seed for reproducibility

# Generate random data for x
x = np.array([random.uniform(0, 10) for _ in range(100)])  # Independent variable (feature)

# Generate corresponding target data for y
y = np.array([2 * xi + random.gauss(0, 1) for xi in x])  # Dependent variable (target)

plt.scatter(x,y)
````

````python
def line_equation(x,w,b):
  '''
  parameters:
  x : numpy array of shape (,100)
  w : the weight of x
  b : the bias

  return:
  y : numpy array of shape (,100)
  '''
  y = x * w + b
  return y
````

````python
LEARNING_RATE = 0.02
ITERATIONS = 10
m = 100 # length of the data
J_list = []

# y = w.x + c
#initialize the values for w and c
w = 0
b = 0

# loop to update the w and c to their optimal vlaues
# updating parameters means GRADIENT DESCENT
for i in range(ITERATIONS):
  y_pred = line_equation(x,w,b)

  # calculate the cost
  J = sum((y - y_pred) ** 2) / m
  print(f"the cost function of iteration {i} is : {J}")
  J_list.append(J)

  # calc the gradients
  dw = - ( 2 * ( x.T ).dot( y - y_pred ) ) / m
  db = - 2 * np.sum( y - y_pred ) / m

  # update the weights
  w = w - dw * LEARNING_RATE
  b = b - db * LEARNING_RATE

y_pred_final =  line_equation(x,w,b)

#plot the data
plt.scatter(x,y)

# plot the line itself
plt.plot(x,y_pred_final, color = 'r')
plt.show()
````

````python
plt.plot(J_list)
````

# Gradient Descent
Gradient descent is an iterative optimization algorithm used to minimize the cost or loss function in machine learning and mathematical optimization. It is commonly used in training models such as linear regression, logistic regression, and neural networks.

The goal of gradient descent is to find the optimal values for the parameters of a model by iteratively updating them in the direction of steepest descent of the cost function. The algorithm starts with initial parameter values and repeatedly adjusts them until it converges to a minimum of the cost function.

Note That Gradient Descent is very dependent on the Cost function you wanna minimize

FOR LINEAR REGRESSION



```
Cost function is MSE : cost J = (y - y_pred)^2
```


so the cost function is of a parabolic shape, has 1 global minmum

note that y_pred is the output of ...


```
y_pred = x * w + b

w and b are called the parameters of the model

So..

cost J = (y - (x * w + b))^2

(x,y) are instances from the data
so, we're left with w and c
the optimal values for w and c determines minimum of the cost

```

repeat until convergence
